```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Reading the dataset
df = pd.read_csv('stock_prices.csv')
df.head()
```

```python
# Converting date column to date time object
df['date'] = pd.to_datetime(df['date'])
df.head()
```

```python
#Setting date as index
df.set_index('date',inplace=True)
```

```python
#Calculating rolling average
rolling_avg = df['closing_price'].rolling(window=30).mean()
```

```python
#plotting Time-Series graph
fig= plt.figure(figsize=(15, 5))
plt.plot(df.index, df['closing_price'], label='Closing Price', color='Red')
plt.plot(rolling_avg.index, rolling_avg, label='30-Day Rolling Avg', color='blue')
plt.title('Stock Prices and 30-Day Rolling Average')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.show()
```
